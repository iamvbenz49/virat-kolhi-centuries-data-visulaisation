```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/viratkohli-76centuries/ViratKohli.76centuries.csv")
df = pd.DataFrame(data)
```

```python
df.index.name = "S.No"
df.index = np.arange(1,77)
```

```python
df.info()
```

```python
missing_count = pd.isnull(df["Strike Rate"]).sum()
missing_count
```

```python
test = df[df["Format"] == "Test"]
test
```

```python
average_test_strike_rate = round(test["Strike Rate"].mean(),2)
average_test_strike_rate
```

```python
df
```

```python
df["Strike Rate"].fillna(average_test_strike_rate, inplace=True)
df
```

```python
df.info()
```

```python
df
```

```python
df["Date"] = df["Date"].str.replace("/", "-")
df["Year"] = df["Date"].str.split("-").str[2]
df
```

```python
centuries_in_year = {}
for year in df["Year"]:
    centuries_in_year.setdefault(int(year),0)
    centuries_in_year[int(year)]+=1
centuries_in_year
```

```python
df["Year"] = df["Year"].astype(int)
centuries_in_year = df.groupby("Year").size()
centuries_in_year= centuries_in_year.to_dict()
for year in range(2009,2024):
    centuries_in_year.setdefault(year,0)
centuries_in_year
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar([i for i in centuries_in_year], [centuries_in_year[i] for i in centuries_in_year])
plt.title("Number of Centuries per Year")
plt.xlabel("Year")
plt.ylabel("Number of Centuries")

plt.show()
plt.savefig('centuries_per_year.png')
```

```python
df
```

```python
batting_order_score = df.loc[:, ['Batting Order', 'Score']]
score = {}
for i in range(1,len(batting_order_score)):
    score.setdefault(batting_order_score.loc[i,'Batting Order'],0)
    batting_order_score.loc[i,'Batting Order']
    score[batting_order_score.loc[i,'Batting Order']]+=batting_order_score.loc[i,'Score']
score
```

```python
batting_order = score.keys()
scores = score.values()


plt.figure(figsize=(10, 6))
plt.bar(batting_order, scores)
plt.title("Runs during at Different batting order")
plt.xlabel("Batting Order")
plt.ylabel("Score")

plt.show()
plt.savefig('centuries_runs_per_batting_order.png')
```
